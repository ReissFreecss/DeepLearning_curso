## **Funciones**
---
Una **función**, en programación, es una sección de código **reutilizable**. Esto quiere decir que puede ser **llamada (o invocada)** por su nombre varias veces, y en cada ocasión realizará la misma acción definida.

Existen dos tipos principales de funciones:
- Las **funciones *built-in*** (integradas), que vienen por defecto en el lenguaje.
- Las **funciones definidas por el usuario**, que puedes crear tú mismo con las herramientas que el lenguaje ofrece.

> 💡 **Nota:** Al llamar a una función, **siempre debe respetarse la sintaxis característica**: el **nombre de la función seguido de paréntesis**.

```python
# Ejemplos de llamadas a funciones:

func()
obtener()
mostrar()
```


Las funciones permiten escribir código más limpio, fácil de leer y mantener. Una de sus mayores ventajas es que podemos reutilizar secciones completas de código cuantas veces lo necesitemos, simplemente llamando a la función por su nombre.

Para crear una función en Python, debemos declararla usando la palabra reservada `def`, seguida del nombre de la función (que sigue las mismas reglas que las variables), y un par de paréntesis que se utilizan para recibir datos externos (parámetros). Todo esto va seguido de dos puntos : que indican el inicio del cuerpo de la función.

````python
def saludar():
    print("¡Hola Mundo!")

def saludar_persona(nombre):
    print(f"Hola, {nombre}!")

saludar()
saludar_persona("A todos")
saludar_persona(pacientes[1])
````

````python
def IMC(estatura: float, peso: int):
  """Calcula el IMC, Estatura en Metros y Peso en Kilos
  estatura float
  peso int
  """
  imc = peso/estatura**2
  print(f"El IMC es {imc:.2f}")
  return imc

# None es la representación existente de la inexistencia misma

imc = IMC(1.80, 80)
imc
````

> 💡**Nota**: Al definir una función, esta no se ejecutará, unicamente será creada para ser llamda en cualquier momento.

### Operadores Booleanos

Los operadores booleanos u operadores lógicos, en el caso de python, no se trata de símbolos, si no, de
palabras pequeñas que tienen la capacidad de hacer comparaciones lógicas con valores booleanos estrictamente.

Estos operadores son regidos por tablas de verdad y reglas muy simples.

#### Tablas de verdad:
---
El resultado de usar el operador `and` es Verdadero solo cuando ambos valores son verdaderos

>| A | B | A and B |
|---|---|---------|
| 1 | 1 | True    |
| 1 | 0 | False   |
| 0 | 1 | False   |
| 0 | 0 | False   |

El resultado de usar el operador `or` es Verdadero cuando al menos uno de los valores es verdadero

>| A | B | A or B |
|---|---|--------|
| 1 | 1 | True   |
| 1 | 0 | True   |
| 0 | 1 | True   |
| 0 | 0 | False  |

El resultado de usar el operador `not`, es el contrario al del valor aplicado, permuta entre 1 y 0

>| A     | not A  |
|-------|--------|
| True  | False  |
| False | True   |


Los operadores booleanos disponibles en python son:
```Python
# Donde A y B son valores lógicos

A and B # operador and
A or B # operador or
not A # operador not
```

**Ejemplos de uso:**
_(Presiona el simbolo de play ▶ a la izquierda para ejecutar la celda)_



````python
A = 7
A > 5 and A < 10 # El resultado es veradero si el valor de A se encuentra entre 5 y 10
````

````python
codon = "ATG"
"T" in codon or "U" in codon # El resultado es verdadero si las letras T o U están presentes en el codon
````

````python
not "ACG" in "ATGTCTAGTAC" # El resultado es verdadero si el codon "ACG" falta en la secuencia
````

````python
A = 7
not (A > 5 and A < 10) # El resultado es veradero si el valor de A no se encuentra entre 5 y 10
````

Una de las ventajas del bloque if-elif-else es la cantidad infinita de preguntas que pueden ser realizadas.

La forma de usar if-elif-else es la siguiente:

````python
secuencia = "ATGTCTAGTAC"

if "U" in secuencia:
    print("ARN")
elif "T" in secuencia:
    print("ADN")
else:
    print("Secuencia inválida")

print("Fin del programa . . .")
````

````python
cali = 100

if cali >= 90:
    if cali >= 95:
        print("Glorioso!")
    else:
      print("Excelente")
elif cali >= 80:
    print("Muy bueno")
elif cali >= 70:
    print("Bueno")
elif cali >= 60:
    print("Regular")
else:
    print("Reprobado")

print("Fin del programa . . .")
````

El código anterior muestra un bloque if-elif-else, el cúal recibe varias condiciones, si el resultado de la comparación
es `True` en la sentencia `if`, entonces el contenido del bloque **`if`** será ejecutado y serán ignoradas el resto de preguntas, en el caso contrario, el programa ejecutará
el bloque `elif`, si la comparación es verdadera, el contenido de **`elif`** será ejecutado y else, ignorado, pero, si nunguna
de las preguntas llega a ser verdadera, entonces **`else`** será ejecutado como última acción.

### Iterativa

Las estructuras de control iterativas, también llamadas bucles, permiten que un programa repita un bloque de código varias veces, de forma automática.

Por ejemplo: si tienes que saludar a 10 personas, en vez de escribir 10 veces print("Hola"), usas un bucle.

#### While
---
El **bucle** `while` es una estructura repetitiva condicional, esto significa que mientras la condición dada sea verdadera,
el contenido de while será ejecutado.

>[![](https://mermaid.ink/img/pako:eNp1kE1OwzAUhK9ivVUruVUS7CT2AommrcSCXVfgLkxsGkNiB5OInzSn4ghcDCcIduzefNLM6M0ApVMaOJy8bCt02Ap7tbi2pjRuiVarS7QZCmdV0F-fdhR2M8Hzwff6jIq73aMu-056dF-7516j18rU-ihs8b91L-uX4N0t9sbKegk4NBsFvAuRGBrtGzlJGIRFSEBX6UYL4OFU0j8JEHYMnlbaW-eaX5t3_akC_jBlY-hbJTu9NTL81PxRr63SvnC97YDn2ZwBfIA34EmWrpOUsJgSklHCUgzvga4pY_Qij1mcJTSmNBsxfMyt0ZoRkkck4JiRKIoSDFqZzvmbnzXnUcdvbo9uPQ?type=png)](https://mermaid.live/edit#pako:eNp1kE1OwzAUhK9ivVUruVUS7CT2AommrcSCXVfgLkxsGkNiB5OInzSn4ghcDCcIduzefNLM6M0ApVMaOJy8bCt02Ap7tbi2pjRuiVarS7QZCmdV0F-fdhR2M8Hzwff6jIq73aMu-056dF-7516j18rU-ihs8b91L-uX4N0t9sbKegk4NBsFvAuRGBrtGzlJGIRFSEBX6UYL4OFU0j8JEHYMnlbaW-eaX5t3_akC_jBlY-hbJTu9NTL81PxRr63SvnC97YDn2ZwBfIA34EmWrpOUsJgSklHCUgzvga4pY_Qij1mcJTSmNBsxfMyt0ZoRkkck4JiRKIoSDFqZzvmbnzXnUcdvbo9uPQ)

El funcionamiento de while implica que si la condición es True todo el tiempo, la cantidad de veces que se ejecute podrían
ser infinitas, llegando a generar pausas indefinidas en el programa, misma razón por la cual se le conoce como **bucle**.


La forma de usar while es la siguiente:




````python
datos = [1, 2, 10, 5, 3]
i = 0
a = 0
n = len(datos)

while i < n:
    print(datos[i])
    i += 1

print("Fin del programa . . .")
````

El código anterior muestra un bloque while, estos bloques se caracterizan por la necesidad de tres variables particulares,
el contador `i`, el limitador `n` y el acumulador `a`; while, al ser un bucle condicional es posible usarlo en diversas
aplicaciones que requieren reintentar el proceso hasta que el resultado deseado no es obtenido, lo cuál puede llevar algunos
pocos o infinitos intentos, en el ejemplo anterior podemos ver el modo "contador" de while, simulando el funcionamiento de
for_in con algunos pasos extras.



````python
while True:
    numA = int(input("Ingrese un número entero: "))
    numB = int(input("Ingrese otro número entero: "))

    print(f"La suma de {numA} + {numB} = {numA + numB}")

    opcion = input("Desea repetir el programa? [Y/N]:  ")

    if opcion == "N":
        break

print("Fin del programa . . .")
````

El ejemplo anterior muestra el uso de while en modo "infinito", esto es cuando se le otorga un valor lógico directamente
en vez de una condición, este modo es utíl para entrar en bucles infinitos controlados desde el interior usando `break` o `continue`.

Además, podemos observar la presencia de las funciones de entrada y salida `E/S - I/O`, `input( )` y `print( )` respectivamente.

Existen palabras reservadas exclusivamente para las estructuras iterativas como:

- **break**
>`break`, es la palabra reservada para "romper" la ejecución de los bloques while y for, para ser
usada solo es necesario buscar la forma de llamarla durante la ejecución del bloque, una forma
sencilla de manejarlo es con un bloque if, si la condición es cumplida, ¡break al programa!

- **continue**
>`continue`, es la palabra reservada para "saltar" vueltas sin detener la ejecución del bloque,
digamos que si cierta condición se cumple no queremos que esa vuelta en particular sea realizada, pero
tampoco que se detenga todo, si no, solo evitar esa ejecución en particular.

#### For in
---
El **ciclo** `for_in`, es una estructura iterativa con una lógica fija, a diferencia del **bucle** `while`, este necesita iterar a través
de una secuencia o colección de datos, que será enviada a su variable auxiliar  un dato a la vez e introducida a la ejecución del bloque for

donde _`i`_ es la variable auxiliar y _`n`_ la cantidad de valores en la secuencia.

La forma de usar for_in es la siguiente:

````python
datos = [1, 2, 10, 5, 3]

for variable_aux in datos:
    print(variable_aux+10)

print("Fin del programa . . .")
````

El código anterior muestra un **ciclo** for, donde la variable auxiliar es `variable_aux` la cual tomará cada uno de los
valores de `datos` a cada "vuelta" y los mostrará por medio de `print( )`

> 💡 **Nota**: Es posible usar como secuencia cualquier tipo de dato que sea de comportamiento
**iterable**, es decir que contenga a más objetos en su interior y pueda ser entrgado cada uno de ellos, por ejemplo:
listas, tuplas, sets, diccionarios, cadenas, etc.

> ⚠️ **Advertencia**: No todos los tipos de datos que almacenan objetos en su interior son iterables por defecto o de manera natural, puede que sea necesaria una función o método para acceder a la parte iterable

El ciclo `for_in` de python es apreciado en diversas áreas, debido a su simplicidad es posible hacer procesamientos extensos
usando solo ciclos for_in

````python
matriz = []
size = 5

for i in range(size):
    matriz.append([])
    for j in range(size):
        matriz[i].append(j+i)

matriz
````


El ejemplo anterior es un programa sencillo con la capacidad de generar matrices númericas cuadradas con tamaño `size`
y valores,.

¿Cómo haríamos si quisiéramos acceder a cada uno de esos valores?

````python
for fila in matriz:
    for col in fila:
        print(col, end=" ")
    print()
````

además de los datos mencionados, existen infinidad de tipos, construidos internamente y otros que pueden ser creados por el programador, cada uno tiene características particulares que los diferencia entre sí, ya sea su comportamiento, capacidades, eficiencia o comodidad.

## Clases

En programación orientada a objetos, una clase es como una plantilla o molde que describe cómo debe ser algo, y un objeto es una instancia concreta creada a partir de esa plantilla.

> Imagina una clase como la raza humana, y un objeto como un individuo de la epecie.

Las clases nos permiten organizar código, agrupar datos y funciones relacionados, y reutilizar estructuras fácilmente.

````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        print(f"Hola, me llamo {self.nombre} y tengo {self.edad} años.")

persona1 = Persona("Ana", 25)
persona2 = Persona("Juan", 30)

persona1.presentarse()
persona2.presentarse()
````

Una clase en Python se compone principalmente de dos elementos:

---

- Atributos  

> Los **atributos** son como las **características o propiedades** de un objeto.  
Se definen generalmente dentro del método `__init__`, y **guardan datos** relevantes para el objeto.

> Por ejemplo: nombre, edad, altura, color, etc.

---

-  Métodos  

> Los **métodos** son las **funciones que están dentro de una clase**.  
Describen el **comportamiento del objeto**, es decir, qué puede hacer.

> Por ejemplo: presentarse, hablar, correr, calcular, etc.
